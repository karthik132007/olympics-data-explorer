```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
import sklearn as sk
```

```python
df=pd.read_csv("data\\athlete_events.csv")
```

```python
# replaced NaN with None
df['Medal']=df['Medal'].replace(np.nan,'None')
```

```python
# Filling missing values in 'Height' and 'Weight' Colmuns
df=df.fillna(df['Height'].mean())
(df['Height'].isnull()).sample(10)
df=df.fillna(df['Weight'].mean())
(df['Height'].isnull()).sample(10)
df[['Height', 'Weight']] = df[['Height', 'Weight']].round(2)
```

```python
#set index as ID
df=df.set_index('ID')
```

```python
# grouped by age 
group_by_sex=df.groupby('Sex')
group_by_sex.agg({
    'Age':['mean','count','median','max','min']
})
```

```python
# created a col is_medlist
df['is_medlist']=df['Medal']!='None'
```

```python
top_countries=df[df['is_medlist']].groupby('Team')
top_countries.size().sort_values(ascending=False).head(10)
```

```python
df=df.drop('Games',axis=1)
```

```python
df=df.drop('NOC',axis=1)
```
